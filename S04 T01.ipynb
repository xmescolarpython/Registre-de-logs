```python
#File: Web_access_log-akumenius.com.txt
#Structure
import numpy as np
import pandas as pd
import seaborn as sns
import requests
import json
import warnings
import os

warnings.filterwarnings('ignore') #Desactivació de Warnings
```

```python
#Preparem una capçalera per les dades del fitxer Web_access_log-akumenius.com.txt
#Identifiquem un reguitzell de dades:
#1.dada de tipus string [localhost]
#2.dada de tipus string [www.akumenius.com]
#3.dada de tipus string - combinació de xifres i simbols  [127.0.0.1 / 66.249.76.216]
#4.dada de tipus date [23/Feb/2014:03:10:31 +0100]
#Es pot dividir en 2 variables:
#4.1.Data [dia/mes/any/hora]
#4.2. Time Zone +0100
#5.dada de tipus string "OPTIONS * HTTP/1.0" / "GET /hoteles-baratos/ofertas-hotel-Club-&-Hotel-Letoonia--en-Fethiye-8460b-destinos.html HTTP/1.1"
#Es pot dividir en 3 variables:
#5.1.GET/POST
#5.2.pàgina htm
#5.3.el tipus de http
#5.dada de tipus númeric 200 / 404
#6.dada de tipus númeric - / 3100
#7.dada de tipus string "Apache (internal dummy connection)" / "Mozilla/5.0 (compatible; Googlebot/2.1; +http://www.google.com/bot.html)"
#Es pot dividir en 4 variables:
#7.1.Software - Apache/Mozilla
#7.2.Description
#7.3.Bot - Googlebot/2.1
#7.4.Connection
#8.data de tipus string [VLOG=-]
#A més a més apareixen diversos simbols que s'hauran de netejar ( - - , "-", ;, ]) 

filePath="Python/Web_access_log.csv"

#Si existeix el fitxer "Python/Web_access_log.csv" l'esborrem.
if os.path.exists(filePath):
    os.remove(filePath)
    
IPfilePath="Python/ips.csv"  

#Si existeix el fitxer "Python/ips.csv" l'esborrem.
if os.path.exists(IPfilePath):
    os.remove(IPfilePath)
    
```

```python
#Neteja, preprocesa, estructura i transforma (dataframe) les dades del registre d'Accés a la web.
#Utilitzem 2 fitxers: 
#fitxer origen: Python/Web_access_log-akumenius.com.txt 
#fitxer destí: Python/Web_access_log.csv - emmagatzematge del dataset net, sense elements de puntuació

with open('Python/Web_access_log.csv', 'a+') as file_a:
    with open('Python/Web_access_log-akumenius.com.txt', 'r') as file_r:
        lines_to_append = file_r.readlines()
        for line in lines_to_append:            
            for character in [' - - [',' "-" "','] "','" ','" ','; ',' (',')"']:
                if character in line:
                   line = line.replace(character,' ')
            file_a.write(line)
    file_r.close()
file_a.close()
```

```python
#Access to new dataframe adding column names
df=pd.read_csv("Python/Web_access_log.csv", header=None, engine="python", error_bad_lines=False, warn_bad_lines=False, sep = ' ')
df.columns=["Web","IP","Created","Code","Comand","Page","Http","Number1","Number2","Browser","Descrip","Bot","Connect","VLOG"]
df.head(15)
```

```python
#Exercici 3
#Geolocalitza les IP's.
#Extracció de les IP's del dataframe
ips = df['IP'].drop_duplicates(keep='last')
ips.tail(10)
```

```python
ips.count # Number of IP's
```

```python
#Geolocalització d'IPs
#Relació d'IPs. 

list_geo_ips = []

for ip in ips: 
    request_url = 'https://geolocation-db.com/jsonp/' + ip
    response = requests.get(request_url) #Recollim els valors de request i reponse
    result = response.content.decode()
    result = result.split("(")[1].strip(")") #Netejem el string rebut
    result  = json.loads(result)
    #Visualitzem les caracteristiques de geolocalització de cada adreça IP
    print('Ip_address: ', ip, '\n ', 
          'Country code: ', result['country_code'], '\n',
          'Country name: ', result['country_name'], '\n',
          'City: ', result['city'], '\n',
          'Postal: ',result['postal'], '\n',
          'Latitude: ', result['latitude'], '\n',
          'Longitude: ', result['longitude'], '\n',
          'State: ', result['state'], '\n')
    list_geo_ips.append(result) #Afegim dades de les IPs
   
```

```python
#Mostrem més dades de les adreces IPs 
#Funció per catalogar la tipologia de l'adreça IP
def classe_ip (num):
    
    if num < 127:
        lletra='A' # Classe A de 1 a 126
    else:
        if num < 192:
            lletra='B' # Classe B de 128 a 191
        else:
            lletra='C' # Classe C de 192 a 223   
    return lletra

for ip in ips: 
    request_url = 'https://geolocation-db.com/jsonp/' + ip
    response = requests.get(request_url) #Recollim els valors de request i reponse
    result = response.content.decode()
    result = result.split("(")[1].strip(")") 
    result  = json.loads(result)
    #Obtenim la primera xifra de l'estructura de l'adreça IP
    num_ip = int(result['IPv4'].replace('.',' ').split()[0])
    #Visualitzem dades de geolocalització
    print(result['country_name'], '\n',
          'Latitude: ', result['latitude'], '/ Longitude: ', result['longitude'], '\n',
          'Ip_address: ', ip, 'Class: ', classe_ip(num_ip),'\n ')
     
```
